```python
data = pd.read_csv("/content/data.csv")
data.head()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
sns.scatterplot(x="radius_mean", y="texture_mean", hue="diagnosis", data=data)
plt.xlabel("Radio Promedio")
plt.ylabel("Textura Promedio")
plt.title("Radio Promedio vs Textura Promedio")

plt.subplot(2, 2, 2)
sns.scatterplot(x="perimeter_mean", y="area_mean", hue="diagnosis", data=data)
plt.xlabel("Perímetro Promedio")
plt.ylabel("Área Promedio")
plt.title("Perímetro Promedio vs Área Promedio")

plt.subplot(2, 2, 3)
sns.scatterplot(x="smoothness_mean", y="compactness_mean", hue="diagnosis", data=data)
plt.xlabel("Suavidad Promedio")
plt.ylabel("Compacidad Promedio")
plt.title("Suavidad Promedio vs Compacidad Promedio")

plt.subplot(2, 2, 4)
sns.scatterplot(x="concavity_mean", y="concave points_mean", hue="diagnosis", data=data)
plt.xlabel("Concavidad Promedio")
plt.ylabel("Puntos Cóncavos Promedio")
plt.title("Concavidad Promedio vs Puntos Cóncavos Promedio")

plt.tight_layout()
plt.show()
```

```python
data['diagnosis_encoded'] = data['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)

# Muestra las primeras filas con la nueva columna
display(data[['diagnosis', 'diagnosis_encoded']].head(20))
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
ax = sns.countplot(x='diagnosis', data=data)
plt.title('Conteo de Diagnósticos')
plt.xlabel('Diagnóstico')
plt.ylabel('Cantidad')

# Calculate percentages
total = len(data)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.show()
```
