#CRISP-DM 1:  **CANCER DE MAMAS**



**¿Qué es el cáncer de mama y por qué es crítico?**

El cáncer de mama es una enfermedad en la que las células del tejido mamario pierden su control de crecimiento y se multiplican de forma anormal, formando tumores que pueden invadir tejidos cercanos y diseminarse a otras partes del cuerpo (metástasis). Puede originarse en los lobulillos (glándulas productoras de leche) o en los conductos (canales que transportan la leche al pezón), siendo el carcinoma ductal invasivo el tipo más común.

El cáncer de mama constituye la enfermedad oncológica más frecuente en mujeres a nivel mundial y una de las principales causas de muerte por cáncer en la población femenina. Su impacto trasciende lo clínico: afecta la salud física, emocional, social y económica de millones de mujeres y sus familias.
Según la Organización Mundial de la Salud (OMS), en 2022 se diagnosticaron más de 2,3 millones de nuevos casos y se registraron aproximadamente 670.000 muertes por esta causa. En Chile, el cáncer de mama representa el 10% de todos los nuevos diagnósticos de cáncer, y en 2023 más de 7.500 mujeres fueron confirmadas con esta enfermedad, siendo el grupo de mayor riesgo el de mujeres mayores de 50 años.

Lo que hace al cáncer de mama especialmente crítico, es su comportamiento silencioso en etapas iniciales. Muchas pacientes no presentan síntomas hasta que la enfermedad ha avanzado, lo que reduce drásticamente las posibilidades de tratamiento curativo. A pesar de los avances en mamografía y programas de detección, la cobertura sigue siendo insuficiente: en Chile, por ejemplo, la cobertura de mamografías en el sistema público fue de apenas un 38,6% en 2022, muy por debajo del estándar recomendado del 70%.

Además, existen brechas significativas entre el sistema público y privado en cuanto a acceso a diagnóstico oportuno, terapias innovadoras y seguimiento post-tratamiento. Estas desigualdades generan un escenario en el que muchas mujeres son diagnosticadas en etapas avanzadas, lo que aumenta la letalidad y los costos del tratamiento, y reduce la calidad de vida.

Desde una perspectiva epidemiológica, el cáncer de mama también representa una carga creciente para los sistemas de salud. Se estima que más del 77% de las mujeres con esta enfermedad fallecen antes de los 80 años, y que la mayor pérdida de años de vida potencial se concentra en mujeres entre 50 y 64 años. Esto convierte al cáncer de mama en un problema prioritario de salud pública, que requiere estrategias integrales de prevención, detección precoz, tratamiento y seguimiento.

En respuesta a esta realidad, organismos como la OMS han lanzado iniciativas globales con metas ambiciosas, como reducir la mortalidad por cáncer de mama en un 2,5% anual, lo que podría evitar más de 2,5 millones de muertes en las próximas décadas. Sin embargo, alcanzar estos objetivos requiere no solo políticas públicas efectivas, sino también una comprensión profunda de los factores clínicos, sociales y estructurales que perpetúan esta enfermedad.


Factores que lo hacen crítico:


*   Alta prevalencia: afecta a 1 de cada 12 mujeres en países con alto desarrollo humano.
*   Impacto emocional y económico: afecta no solo a la paciente, sino también a su entorno familiar y laboral.

*   Desigualdad en el acceso al diagnóstico: en países con menos recursos, muchas mujeres son diagnosticadas en etapas avanzadas, lo que reduce las posibilidades de tratamiento efectivo.
*   Necesidad de diagnóstico temprano: cuando se detecta en etapas iniciales, la tasa de supervivencia a 5 años puede superar el 90%.

Impacto Global


*   Es el cáncer más diagnosticado en mujeres en 157 de 185 países.
*   En países con bajo índice de desarrollo humano, la mortalidad es proporcionalmente más alta debido a la falta de acceso a diagnóstico precoz y tratamiento oportuno.
*   La detección temprana puede reducir la mortalidad hasta en un 30%, según la OMS.
¿Cómo se detecta?
El diagnóstico del cáncer de mama se basa en una combinación de métodos clínicos, imagenológicos y citológicos:


*   Mamografía: técnica de imagen utilizada para el cribado poblacional. Permite detectar lesiones no palpables.
*   Ecografía mamaria: útil para caracterizar nódulos y guiar biopsias.
*   Biopsia por aspiración con aguja fina (FNA): permite obtener células del tumor para su análisis morfológico.
*   Análisis citológico y patológico: determinan si las células son benignas o malignas, y en qué grado.


#CRISP-DM 2: Conocer los datos

En este proyecto, se trabaja con un dataset derivado de los resultados del exámen FNA, donde se han extraído 30 variables morfológicas que describen características como:

*   Textura (texture_mean, texture_worst)
*   Simetría y complejidad (symmetry_mean, fractal_dimension_worst)



*   Target (**diagnosis**): Esta columna nos muestra el resultado de los análisis clínicos, y es la variable objetivo a evaluar.
   
    *   M = Maligo
    *   B = Benigno

**Para las siguientes columnas no existe un rango universal fijo que determine con certeza si un valor es maligno o benigno, ya que depende del conjunto de datos y del modelo utilizado.**


*   **radius_mean** : Representa el promedio del radio de los núcleos celulares de un tumor mamario. Mide la distancia media desde el centro hasta su borde. Los tumores malignos tienden a tener núcleos más grandes y de forma más irregular generalmente con valores entre 15 y 28.


*   **texture_mean** : Representa el promedio de la variación en la intensidad de los píxeles dentro del núcleo celular. Describe cuán rugosa o granulada se ve la textura. Los tumores malignos tienden a tener texturas más irregulares o heterogéneas. Un valor alto podría estar asociado con mayor probabilidad de malignidad generalmente con valores entre 15 y 30.


*   **perimeter_mean**: Representa el promedio del perímetro del núcleo celular. mide la longitud del contorno en cada célula y luego se promedia para cada muestra. Tumores malignos suelen tener núcleos más grandes y con contornos más irregulares, generalmente con valores entre 90 y 180.

*   **area_mean**: Representa el área promedio del núcleo celular. Mide el tamaño proyectado en dos dimensiones. Los tumores malignos suelen tener núcleos más grandes y con formas irregulares. Esta variable es un buen indicador del crecimiento celular anormal, generalmente con valores entre 600 y 2500.

*   **smoothness_mean**: Mide la variación local en la longitud del radio del núcleo celular. Evalúa cuán suave o irregular es el contorno. Un valor alto sugiere un contorno más irregular o dentado, lo que puede ser característico de tumores malignos, donde las células se dividen de forma desorganizada.Generalmente con valores entre 0.10 y 0.16.


*   **compactness_mean**: Mide el grado de compacidad de los bordes del núcleo celular, considerando la relación entre el perímetro y el área. Evalúa cuán ajustados o extendidos son los contornos del tumor. Un valor alto indica bordes más irregulares y extendidos, lo que suele relacionarse con tumores malignos que crecen de forma desorganizada, generalmente con valores entre 0.02 y 0.15.

*   **concavity_mean**: Mide el promedio de la severidad de las partes cóncavas del contorno del núcleo celular. Evalúa qué tan profundas o marcadas son las hendiduras en el borde. Las células malignas tienden a tener contornos más irregulares y con más hendiduras pronunciadas, generalmente con valores entre 0.10 y 0.20.

*   **concave points_mean**: Mide el promedio del número de puntos cóncavos en el contorno del núcleo celular. Cuenta cuántas hendiduras o “entradas” hay en el borde. Los tumores malignos suelen tener contornos más irregulares, con más hendiduras profundas y numerosas, generalmente con valores entre 0.05 y 0.20.

*   **symmetry_mean**: Mide el promedio de la simetría del núcleo celular, es decir, cuán similar es un lado del núcleo respecto al otro. Se calcula a partir de la diferencia entre pares de puntos opuestos en el contorno, evaluando su regularidad. Los tumores malignos suelen presentar asimetrías marcadas debido al crecimiento celular desorganizado, generalmente con valores entre 0.20 y 0.35.

*   **fractal_dimension_mean**: Mide el promedio de la dimensión fractal del contorno del núcleo celular. Evalúa la complejidad geométrica del borde, es decir, cuán irregular o intrincado es su contorno. Cuanto más irregular y complejo es el borde del núcleo, mayor es su dimensión fractal, generalmente con valores entre 0.06 y 0.10.

*   **radius_se**: Representa el error estándar del radio del núcleo celular, mide cuánto varía el radio de los núcleos celulares dentro de una misma muestra. Un valor alto sugiere mayor variabilidad, lo que puede ser un signo de heterogeneidad celular común en tumores malignos, generalmente con valores entre 0.3 y 1.5.

*   **texture_se**: Representa el error estándar de la textura del núcleo celular. Mide cuánto varía la textura (rugosidad o granularidad) entre las distintas células de una misma muestra. Un valor alto sugiere mayor variabilidad en la textura, lo que puede ser un signo de heterogeneidad celular común en tumores malignos, generalmente con valores entre 1.0 y 2.5.

*   **perimeter_se**: Representa el error estándar del perímetro del núcleo celular. Mide cuánto varía la longitud del contorno de los núcleos celulares dentro de una misma muestra. Un valor alto sugiere mayor variabilidad, lo que puede reflejar heterogeneidad celular, una característica común en tumores malignos, generalmente con valores entre 1.5 y 5.0.

*   **area_se**: Representa el error estándar del área del núcleo celular. Mide cuánto varía el tamaño del área de los núcleos celulares dentro de una misma muestra. Un valor alto sugiere mayor variabilidad en el tamaño, lo que puede reflejar heterogeneidad celular, una característica común en tumores malignos, generalmente con valores entre 40 y 200.

*   **smoothness_se**: Representa el error estándar de la suavidad del contorno del núcleo celular. Mide cuánto varía la suavidad entre los distintos núcleos dentro de una misma muestra. Un valor alto sugiere que hay variabilidad en la regularidad del contorno, lo que puede reflejar desorganización celular común en tumores malignos, generalmente con valores entre 0.005 y 0.015.

*   **compactness_se**: Representa el error estándar de la compacidad del núcleo celular.  Mide cuánto varía esa relación entre perímetro y área entre los distintos núcleos de una misma muestra. Un valor alto sugiere variabilidad en la forma o densidad de los núcleos, lo que puede ser un signo de desorganización celular, típico en tumores malignos, generalmente con valores entre 0.02 y 0.10.

*   **concavity_se**: Representa el error estándar de la concavidad del núcleo celular. Mide cuánto varía la profundidad de las hendiduras en los bordes de los núcleos dentro de una misma muestra. Un valor alto sugiere variabilidad en la forma de los bordes, lo que puede reflejar desorganización celular típica en tumores malignos, generalmente con valores entre 0.02 y 0.15.

*   **concave points_se**: Representa el error estándar del número de puntos cóncavos en el contorno del núcleo celular. Mide cuánto varía la cantidad de hendiduras en los bordes de los núcleos dentro de una misma muestra. Un valor alto sugiere variabilidad en la cantidad de concavidades, lo que puede reflejar heterogeneidad estructural típica en tumores malignos, generalmente con valores entre 0.01 y 0.05.

*   **symmetry_se**: Representa el error estándar de la simetría del núcleo celular. Mide cuánto varía la simetría entre los distintos núcleos dentro de una misma muestra. Un valor alto sugiere variabilidad en la simetría, lo que puede reflejar desorganización estructural común en tumores malignos, generalmente con valores entre 0.02 y 0.08.

*   **fractal_dimension_se**: Representa el error estándar de la dimensión fractal del contorno del núcleo celular. Mide cuánto varía la complejidad del borde de los núcleos dentro de una misma muestra. Un valor alto sugiere variabilidad en la irregularidad del borde, lo que puede reflejar desorganización estructural, típica en tejidos malignos, generalmente con valores entre 0.005 y 0.03.

*   **radius_worst**: Representa el promedio de los tres valores más altos del radio del núcleo celular en una muestra. Mide el tamaño más grande observado entre los núcleos, calculado como la media de los tres radios más grandes.
Un valor alto de radius_worst indica que hay núcleos con radios significativamente grandes, lo que puede ser un signo de crecimiento celular anormal, típico en tumores malignos, generalmente con valores entre 15 y 36.

*   **texture_worst**: Representa el valor más alto (promedio de los tres valores más altos) de la textura del núcleo celular en una muestra. Un valor alto indica que al menos tres núcleos en la muestra tienen una textura muy irregular o compleja. Esto puede ser un indicador fuerte de malignidad, ya que los tumores malignos tienden a mostrar mayor heterogeneidad celular, generalmente con valores entre 20 y 40.

*   **perimeter_worst**: Representa el promedio de los tres perímetros más grandes del núcleo celular en una muestra. Un valor alto indica que hay núcleos con contornos muy extensos, lo que puede reflejar crecimiento celular anormal y ser un fuerte indicio de malignidad, generalmente con valores entre 100 y 250.

*   **area_worst**: Representa el promedio de las tres áreas más grandes del núcleo celular en una muestra. Mide el tamaño más extremo observado entre los núcleos. Un valor alto indica que hay núcleos con áreas significativamente grandes, lo cual puede ser un fuerte indicio de crecimiento celular anormal, típico en tumores malignos, generalmente con valores entre 800 y 2500.

*   **smoothness_worst**: Representa el promedio de la suavidad del contorno de los tres núcleos celulares más irregulares de una muestra. Mide la mayor irregularidad observada en los bordes de los núcleos. Un valor alto sugiere que hay núcleos con bordes muy irregulares o dentados, lo cual es típico en tumores malignos, generalmente con valores entre 0.12 y 0.25.

*   **compactness_worst**: Representa el promedio de los tres valores más altos de la compacidad del núcleo celular en una muestra. Un valor alto indica que hay núcleos con formas muy irregulares o extendidas, lo cual puede ser un fuerte indicio de malignidad, generalmente con valores entre 0.15 y 0.60.

*   **concavity_worst**: Representa el promedio de las tres concavidades más profundas del contorno del núcleo celular en una muestra. Mide la severidad máxima de las hendiduras en los bordes de los núcleos. Un valor alto de concavity_worst indica que hay núcleos con hendiduras muy marcadas, lo que puede ser un fuerte indicio de malignidad, generalmente con valores entre 0.20 y 1.20

*   **concave points_worst**: Representa el promedio de los tres valores más altos del número de puntos cóncavos en el contorno del núcleo celular. Mide cuántas hendiduras profundas y marcadas aparecen en los bordes de los núcleos más irregulares de una muestra. Un valor alto indica que hay núcleos con muchos puntos cóncavos, lo que sugiere alta irregularidad estructural, típica en tumores malignos, generalmente con valores entre 0.10 y 0.40.

*   **symmetry_worst**: Representa el promedio de los tres valores más altos de simetría del núcleo celular en una muestra. Mide la mayor asimetría observada entre los núcleos, calculada a partir de los tres núcleos más irregulares.
Un valor alto indica que hay núcleos con formas muy asimétricas, lo que puede ser un fuerte indicio de malignidad, generalmente con valores entre 0.30 y 0.60.

*   **fractal_dimension_worst**: Representa el promedio de los tres valores más altos de la dimensión fractal del contorno del núcleo celular en una muestra. La dimensión fractal es una medida matemática que describe cuán complejo o irregular es el borde del núcleo, capturando patrones que no son evidentes a simple vista. Un valor alto sugiere que hay núcleos con contornos altamente complejos, lo que puede reflejar desorganización celular extrema, típica en tumores malignos avanzados, generalmente con valores entre 0.07 y 0.20.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Lista de variables importantes para graficar
variables_outlier = ['radius_worst', 'concavity_worst', 'area_mean', 'perimeter_worst']

plt.figure(figsize=(6, 8))
for i, var in enumerate(variables_outlier):
    plt.subplot(2, 2, i+1)
    sns.boxplot(y=data[var], color='lightsteelblue')  # ¡Aquí cambiamos de x= a y=!
    plt.title(f'Boxplot de {var}')
plt.tight_layout()
plt.show()
```

4. Boxplot de perimeter_worst : Muestra el perímetro más grande observado del tumor. Aquí también se ve una cola larga hacia la derecha, indicando valores extremos bien diferenciados.

```python
#Identificar los registros que son outliers con casos extremos de radius_worst:

# Calcular umbrales para detectar outliers por IQR
Q1 = data['radius_worst'].quantile(0.25)
Q3 = data['radius_worst'].quantile(0.75)
IQR = Q3 - Q1


# Filtrar registros que están fuera de ese rango
outliers_radius = data[(data['radius_worst'] < Q1 - 1.5 * IQR) | (data['radius_worst'] > Q3 + 1.5 * IQR)]

# Mostrar los registros
display(outliers_radius[['diagnosis', 'radius_worst']])

```

**Esta tabla muestra los valores atípicos detectados, junto con su correspondiente diagnóstico (M o B).**

Ese bloque de código identifica y muestra los outliers (valores atípicos) en la columna radius_worst,
es decir, aquellos tumores cuyo radio máximo se considera extremadamente bajo o alto en comparación  con el resto del conjunto de datos.

Se filtran los registros que están fuera de ese rango, es decir, tumores cuyo radius_worst es excesivamente pequeño o grande.

Este análisis  ayuda a detectar tumores que sobresalen en tamaño, lo que puede ser útil para entender variaciones morfológicas extremas en la muestra y  decidir si conviene tratarlos como casos especiales.

#2. ¿Qué proporción de tumores benignos y malignos presenta una concavidad extrema (concavity_worst > 0.4)?

```python
# Crear nueva columna booleana según umbral crítico de concavidad
data['umbral_concavidad'] = data['concavity_worst'] > 0.4

# Agrupar por diagnóstico y si supera el umbral
concavidad_data = data.groupby(['diagnosis', 'umbral_concavidad']).size().unstack().fillna(0)

# Calcular porcentaje para representación visual
concavidad_percent = concavidad_data.div(concavidad_data.sum(axis=1), axis=0) * 100

# Crear gráfico de barras apiladas
concavidad_percent.plot(kind='bar', stacked=True, figsize=(6, 4), color=['lightblue', 'orange'])
plt.title('Proporción de tumores con concavidad extrema (> 0.4)')
plt.xlabel('Diagnóstico')
plt.ylabel('Porcentaje (%)')
plt.legend(['≤ 0.4', '> 0.4'], title='Concavidad Peor Caso')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Este gráfico revela qué tan común es encontrar concavidades profundas entre tumores benignos y malignos. Como era de esperar, la gran mayoría de los tumores malignos presenta valores de concavity_worst superiores a 0.4 (área naranja), mientras que los benignos tienden a ubicarse por debajo de ese umbral (área celeste). Esta separación visual valida que el nivel de concavidad es un indicador confiable de malignidad, lo cual es clave para apoyar sistemas de diagnóstico temprano o modelos predictivos.
