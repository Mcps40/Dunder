En este proyecto, se trabaja con un dataset derivado de los resultados del exmen FNA, donde se han extrado 30 variables morfolgicas que describen caractersticas como:

*   Target (**diagnosis**): Esta columna nos muestra el resultado de los anlisis clnicos, y es la variable objetivo a evaluar.
   
    *   M = Maligo
    *   B = Benigno

**Para las siguientes columnas no existe un rango universal fijo que determine con certeza si un valor es maligno o benigno, ya que depende del conjunto de datos y del modelo utilizado.**


*   **radius_mean** : Representa el promedio del radio de los ncleos celulares de un tumor mamario. Mide la distancia media desde el centro hasta su borde. Los tumores malignos tienden a tener ncleos ms grandes y de forma ms irregular generalmente con valores entre 15 y 28.


*   **texture_mean** : Representa el promedio de la variacin en la intensidad de los pxeles dentro del ncleo celular. Describe cun rugosa o granulada se ve la textura. Los tumores malignos tienden a tener texturas ms irregulares o heterogneas. Un valor alto podra estar asociado con mayor probabilidad de malignidad generalmente con valores entre 15 y 30.


*   **perimeter_mean**: Representa el promedio del permetro del ncleo celular. mide la longitud del contorno en cada clula y luego se promedia para cada muestra. Tumores malignos suelen tener ncleos ms grandes y con contornos ms irregulares, generalmente con valores entre 90 y 180.

*   **area_mean**: Representa el rea promedio del ncleo celular. Mide el tamao proyectado en dos dimensiones. Los tumores malignos suelen tener ncleos ms grandes y con formas irregulares. Esta variable es un buen indicador del crecimiento celular anormal, generalmente con valores entre 600 y 2500.

*   **smoothness_mean**: Mide la variacin local en la longitud del radio del ncleo celular. Evala cun suave o irregular es el contorno. Un valor alto sugiere un contorno ms irregular o dentado, lo que puede ser caracterstico de tumores malignos, donde las clulas se dividen de forma desorganizada.Generalmente con valores entre 0.10 y 0.16.


*   **compactness_mean**: Mide el grado de compacidad de los bordes del ncleo celular, considerando la relacin entre el permetro y el rea. Evala cun ajustados o extendidos son los contornos del tumor. Un valor alto indica bordes ms irregulares y extendidos, lo que suele relacionarse con tumores malignos que crecen de forma desorganizada, generalmente con valores entre 0.02 y 0.15.

*   **concavity_mean**: Mide el promedio de la severidad de las partes cncavas del contorno del ncleo celular. Evala qu tan profundas o marcadas son las hendiduras en el borde. Las clulas malignas tienden a tener contornos ms irregulares y con ms hendiduras pronunciadas, generalmente con valores entre 0.10 y 0.20.

*   **concave points_mean**: Mide el promedio del nmero de puntos cncavos en el contorno del ncleo celular. Cuenta cuntas hendiduras o entradas hay en el borde. Los tumores malignos suelen tener contornos ms irregulares, con ms hendiduras profundas y numerosas, generalmente con valores entre 0.05 y 0.20.

*   **symmetry_mean**: Mide el promedio de la simetra del ncleo celular, es decir, cun similar es un lado del ncleo respecto al otro. Se calcula a partir de la diferencia entre pares de puntos opuestos en el contorno, evaluando su regularidad. Los tumores malignos suelen presentar asimetras marcadas debido al crecimiento celular desorganizado, generalmente con valores entre 0.20 y 0.35.

*   **fractal_dimension_mean**: Mide el promedio de la dimensin fractal del contorno del ncleo celular. Evala la complejidad geomtrica del borde, es decir, cun irregular o intrincado es su contorno. Cuanto ms irregular y complejo es el borde del ncleo, mayor es su dimensin fractal, generalmente con valores entre 0.06 y 0.10.

*   **radius_se**: Representa el error estndar del radio del ncleo celular, mide cunto vara el radio de los ncleos celulares dentro de una misma muestra. Un valor alto sugiere mayor variabilidad, lo que puede ser un signo de heterogeneidad celular comn en tumores malignos, generalmente con valores entre 0.3 y 1.5.

*   **texture_se**: Representa el error estndar de la textura del ncleo celular. Mide cunto vara la textura (rugosidad o granularidad) entre las distintas clulas de una misma muestra. Un valor alto sugiere mayor variabilidad en la textura, lo que puede ser un signo de heterogeneidad celular comn en tumores malignos, generalmente con valores entre 1.0 y 2.5.

*   **perimeter_se**: Representa el error estndar del permetro del ncleo celular. Mide cunto vara la longitud del contorno de los ncleos celulares dentro de una misma muestra. Un valor alto sugiere mayor variabilidad, lo que puede reflejar heterogeneidad celular, una caracterstica comn en tumores malignos, generalmente con valores entre 1.5 y 5.0.

*   **area_se**: Representa el error estndar del rea del ncleo celular. Mide cunto vara el tamao del rea de los ncleos celulares dentro de una misma muestra. Un valor alto sugiere mayor variabilidad en el tamao, lo que puede reflejar heterogeneidad celular, una caracterstica comn en tumores malignos, generalmente con valores entre 40 y 200.

*   **smoothness_se**: Representa el error estndar de la suavidad del contorno del ncleo celular. Mide cunto vara la suavidad entre los distintos ncleos dentro de una misma muestra. Un valor alto sugiere que hay variabilidad en la regularidad del contorno, lo que puede reflejar desorganizacin celular comn en tumores malignos, generalmente con valores entre 0.005 y 0.015.

*   **compactness_se**: Representa el error estndar de la compacidad del ncleo celular.  Mide cunto vara esa relacin entre permetro y rea entre los distintos ncleos de una misma muestra. Un valor alto sugiere variabilidad en la forma o densidad de los ncleos, lo que puede ser un signo de desorganizacin celular, tpico en tumores malignos, generalmente con valores entre 0.02 y 0.10.

*   **concavity_se**: Representa el error estndar de la concavidad del ncleo celular. Mide cunto vara la profundidad de las hendiduras en los bordes de los ncleos dentro de una misma muestra. Un valor alto sugiere variabilidad en la forma de los bordes, lo que puede reflejar desorganizacin celular tpica en tumores malignos, generalmente con valores entre 0.02 y 0.15.

*   **concave points_se**: Representa el error estndar del nmero de puntos cncavos en el contorno del ncleo celular. Mide cunto vara la cantidad de hendiduras en los bordes de los ncleos dentro de una misma muestra. Un valor alto sugiere variabilidad en la cantidad de concavidades, lo que puede reflejar heterogeneidad estructural tpica en tumores malignos, generalmente con valores entre 0.01 y 0.05.

*   **symmetry_se**: Representa el error estndar de la simetra del ncleo celular. Mide cunto vara la simetra entre los distintos ncleos dentro de una misma muestra. Un valor alto sugiere variabilidad en la simetra, lo que puede reflejar desorganizacin estructural comn en tumores malignos, generalmente con valores entre 0.02 y 0.08.

*   **fractal_dimension_se**: Representa el error estndar de la dimensin fractal del contorno del ncleo celular. Mide cunto vara la complejidad del borde de los ncleos dentro de una misma muestra. Un valor alto sugiere variabilidad en la irregularidad del borde, lo que puede reflejar desorganizacin estructural, tpica en tejidos malignos, generalmente con valores entre 0.005 y 0.03.

*   **radius_worst**: Representa el promedio de los tres valores ms altos del radio del ncleo celular en una muestra. Mide el tamao ms grande observado entre los ncleos, calculado como la media de los tres radios ms grandes.
Un valor alto de radius_worst indica que hay ncleos con radios significativamente grandes, lo que puede ser un signo de crecimiento celular anormal, tpico en tumores malignos, generalmente con valores entre 15 y 36.

*   **texture_worst**: Representa el valor ms alto (promedio de los tres valores ms altos) de la textura del ncleo celular en una muestra. Un valor alto indica que al menos tres ncleos en la muestra tienen una textura muy irregular o compleja. Esto puede ser un indicador fuerte de malignidad, ya que los tumores malignos tienden a mostrar mayor heterogeneidad celular, generalmente con valores entre 20 y 40.

*   **perimeter_worst**: Representa el promedio de los tres permetros ms grandes del ncleo celular en una muestra. Un valor alto indica que hay ncleos con contornos muy extensos, lo que puede reflejar crecimiento celular anormal y ser un fuerte indicio de malignidad, generalmente con valores entre 100 y 250.

*   **area_worst**: Representa el promedio de las tres reas ms grandes del ncleo celular en una muestra. Mide el tamao ms extremo observado entre los ncleos. Un valor alto indica que hay ncleos con reas significativamente grandes, lo cual puede ser un fuerte indicio de crecimiento celular anormal, tpico en tumores malignos, generalmente con valores entre 800 y 2500.

*   **smoothness_worst**: Representa el promedio de la suavidad del contorno de los tres ncleos celulares ms irregulares de una muestra. Mide la mayor irregularidad observada en los bordes de los ncleos. Un valor alto sugiere que hay ncleos con bordes muy irregulares o dentados, lo cual es tpico en tumores malignos, generalmente con valores entre 0.12 y 0.25.

*   **compactness_worst**: Representa el promedio de los tres valores ms altos de la compacidad del ncleo celular en una muestra. Un valor alto indica que hay ncleos con formas muy irregulares o extendidas, lo cual puede ser un fuerte indicio de malignidad, generalmente con valores entre 0.15 y 0.60.

*   **concavity_worst**: Representa el promedio de las tres concavidades ms profundas del contorno del ncleo celular en una muestra. Mide la severidad mxima de las hendiduras en los bordes de los ncleos. Un valor alto de concavity_worst indica que hay ncleos con hendiduras muy marcadas, lo que puede ser un fuerte indicio de malignidad, generalmente con valores entre 0.20 y 1.20

*   **concave points_worst**: Representa el promedio de los tres valores ms altos del nmero de puntos cncavos en el contorno del ncleo celular. Mide cuntas hendiduras profundas y marcadas aparecen en los bordes de los ncleos ms irregulares de una muestra. Un valor alto indica que hay ncleos con muchos puntos cncavos, lo que sugiere alta irregularidad estructural, tpica en tumores malignos, generalmente con valores entre 0.10 y 0.40.

*   **symmetry_worst**: Representa el promedio de los tres valores ms altos de simetra del ncleo celular en una muestra. Mide la mayor asimetra observada entre los ncleos, calculada a partir de los tres ncleos ms irregulares.
Un valor alto indica que hay ncleos con formas muy asimtricas, lo que puede ser un fuerte indicio de malignidad, generalmente con valores entre 0.30 y 0.60.

*   **fractal_dimension_worst**: Representa el promedio de los tres valores ms altos de la dimensin fractal del contorno del ncleo celular en una muestra. La dimensin fractal es una medida matemtica que describe cun complejo o irregular es el borde del ncleo, capturando patrones que no son evidentes a simple vista. Un valor alto sugiere que hay ncleos con contornos altamente complejos, lo que puede reflejar desorganizacin celular extrema, tpica en tumores malignos avanzados, generalmente con valores entre 0.07 y 0.20.

```python
#Identificar los registros que son outliers con casos extremos de radius_worst:

# Calcular umbrales para detectar outliers por IQR
Q1 = data['radius_worst'].quantile(0.25)
Q3 = data['radius_worst'].quantile(0.75)
IQR = Q3 - Q1


# Filtrar registros que estn fuera de ese rango
outliers_radius = data[(data['radius_worst'] < Q1 - 1.5 * IQR) | (data['radius_worst'] > Q3 + 1.5 * IQR)]

# Mostrar los registros
display(outliers_radius[['diagnosis', 'radius_worst']])

```

**Esta tabla muestra los valores atpicos detectados, junto con su correspondiente diagnstico (M o B).**

Ese bloque de cdigo identifica y muestra los outliers (valores atpicos) en la columna radius_worst,
es decir, aquellos tumores cuyo radio mximo se considera extremadamente bajo o alto en comparacin  con el resto del conjunto de datos.

Se filtran los registros que estn fuera de ese rango, es decir, tumores cuyo radius_worst es excesivamente pequeo o grande.

Este anlisis  ayuda a detectar tumores que sobresalen en tamao, lo que puede ser til para entender variaciones morfolgicas extremas en la muestra y  decidir si conviene tratarlos como casos especiales.

#2. Qu proporcin de tumores benignos y malignos presenta una concavidad extrema (concavity_worst > 0.4)?

```python
# Crear nueva columna booleana segn umbral crtico de concavidad
data['umbral_concavidad'] = data['concavity_worst'] > 0.4

# Agrupar por diagnstico y si supera el umbral
concavidad_data = data.groupby(['diagnosis', 'umbral_concavidad']).size().unstack().fillna(0)

# Calcular porcentaje para representacin visual
concavidad_percent = concavidad_data.div(concavidad_data.sum(axis=1), axis=0) * 100

# Crear grfico de barras apiladas
concavidad_percent.plot(kind='bar', stacked=True, figsize=(6, 4), color=['lightblue', 'orange'])
plt.title('Proporcin de tumores con concavidad extrema (> 0.4)')
plt.xlabel('Diagnstico')
plt.ylabel('Porcentaje (%)')
plt.legend([' 0.4', '> 0.4'], title='Concavidad Peor Caso')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Este grfico revela qu tan comn es encontrar concavidades profundas entre tumores benignos y malignos. Como era de esperar, la gran mayora de los tumores malignos presenta valores de concavity_worst superiores a 0.4 (rea naranja), mientras que los benignos tienden a ubicarse por debajo de ese umbral (rea celeste).
